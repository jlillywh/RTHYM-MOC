## 2. Standpipe size sweep

```python
from surge_design_rules_verification_utils import (
    evaluate_standpipe_sweep,
    standpipe_sweep_monotonic,
)

standpipe = evaluate_standpipe_sweep()
areas = [p.area_ft2 for p in standpipe]
peaks = [p.peak_head_ft for p in standpipe]
limits = [p.peak_limit_ft for p in standpipe]
print(f"Monotonic: {standpipe_sweep_monotonic(standpipe)}")
for p in standpipe:
    print(f"  {p.area_ft2:4.0f} ft²: peak={p.peak_head_ft:.1f} ft limit={p.peak_limit_ft:.0f} PASS={p.passed}")

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(areas, peaks, "o-", label="Simulated peak")
ax.plot(areas, limits, "s--", color="gray", label="pytest upper bound")
ax.set_xlabel("Standpipe area (ft²)")
ax.set_ylabel("SP1 peak head (ft)")
ax.set_title("Closure peak vs standpipe size")
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

## 3. Hydropneumatic placement

```python
from surge_design_rules_verification_utils import evaluate_placement_sweep

placement = evaluate_placement_sweep()
dists = [p.distance_ft for p in placement]
trip_h = [p.trip_mean_head_ft for p in placement]
for p in placement:
    print(f"  {p.distance_ft:.0f} ft: trip mean={p.trip_mean_head_ft:.1f} ft floor={p.trip_head_floor_ft:.0f} PASS={p.passed}")

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dists, trip_h, "o-", color="darkcyan")
ax.set_xlabel("Distance from pump to HPT (ft)")
ax.set_ylabel("Trip-window mean head at Jd (ft)")
ax.set_title("Protection weakens as vessel moves downstream")
ax.grid(True, alpha=0.3)
plt.show()
```
