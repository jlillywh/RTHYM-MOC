## 1. Setup

```python
import matplotlib.pyplot as plt
import numpy as np
import rthym_moc
from _verification_notebook_setup import bootstrap_verification_notebook

REPO_ROOT, TESTS_DIR = bootstrap_verification_notebook()
print(f"Repository root: {REPO_ROOT}")
print(f"rthym_moc: {getattr(rthym_moc, '__version__', 'unknown')}")
```

## 2. Run sweep

```python
from gradual_closure_verification_utils import (
    JOUKOWSKY_DH_FT,
    T_WAVE_S,
    run_all_closure_cases,
)

cases = run_all_closure_cases()
print(f"2L/a = {T_WAVE_S:.2f} s, Joukowsky dH = {JOUKOWSKY_DH_FT:.1f} ft")
for c in cases:
    print(f"  {c.label}: fraction={c.observed_fraction:.3f} PASS={c.passed}")
```

## 3. Valve head traces

```python
fig, axes = plt.subplots(len(cases), 1, figsize=(12, 3 * len(cases)), sharex=False)
if len(cases) == 1:
    axes = [axes]
for ax, c in zip(axes, cases):
    ax.plot(c.time_s, c.valve_head_ft, label="V1 head")
    ax.axhline(c.observed_dh_ft + c.valve_head_ft.min(), color="gray", linestyle=":", alpha=0.5)
    ax.set_title(f"{c.label}: dH fraction {c.observed_fraction:.2f} (PASS={c.passed})")
    ax.set_ylabel("Head (ft)")
    ax.grid(True, alpha=0.3)
axes[-1].set_xlabel("Time (s)")
fig.tight_layout()
plt.show()
```
